### Operator Precedence

#### Operators                                                        Description
##### ()                                                             Parenthesis
##### **                                                                Power
##### * / // %                                        Multiplication, Division, Floor Division, Modulus
##### + -                                                        Addition, Subtraction 

```python
a = 5
b = 2
a=b
print(a)
```

```python
a = 5
b = 2
a=a+b
print(a)
```

```python
a = 5 
b = 2
a=a-b
print(a)
```

```python
a = 5 
b = 2
a=a*b
print(a)
```

```python
a = 5 
b = 2
a=a/b
print(a)
```

```python
a = 5 
b = 2
a=a//b
print(a)
```

```python
a = 5 
b = 2
a=a%b
print(a)
```

```python
a = 5 
b = 2
a=a**b
print(a)
```
