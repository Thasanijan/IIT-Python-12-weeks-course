## Comparison Operators

### a==b (equal)
### False

### a!=b (Not equal)
### True

### a>b  (a greater than b)
### True

### a<b  (b greater than a)
### False

### a>=b (a greater than and equals b)
### True

### a<=b (b greater than and equals a)
### False
