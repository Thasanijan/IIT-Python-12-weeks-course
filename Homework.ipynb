```python
# Prompting the user for input
name = input("Enter your name: ")
marks = float(input("Enter your marks: "))
```

```python
if marks > 80:
    grade = "A"
elif marks >= 60:
    grade = "B"
elif marks >= 40:
    grade = "C"
else:
    grade = "Fail"
```

```python
# Displaying the result
print("----- Student Result -----")
print("Name :", name)
print("Marks:", marks)
print("Grade:", grade)
```
